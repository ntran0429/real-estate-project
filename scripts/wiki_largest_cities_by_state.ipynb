```python
import pandas as pd
import html5lib
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_cities_of_U.S._states_and_territories_by_population'

tables = pd.read_html(url)
```

```python
largest_cities = tables[2]
```

```python
# rename columns of the table
largest_cities.columns = ['State', 'State Population', 'City1', 'Population1', 'City2', 'Population2'
                          , 'City3', 'Population3', 'City4', 'Population4', 'City5', 'Population5'
                          , 'Capital', 'Capital Population']
```

```python
largest_cities.head(10)
```

```python
# remove first row
largest_cities = largest_cities.iloc[1:]
```

```python
largest_cities.head(25)
```

```python
largest_cities['State'] = largest_cities['State'].str.replace(r'\d$', '', regex=True)
```

```python
# output to csv
largest_cities.to_csv('../raw data/largest_cities.csv', index=False)
```

```python
# select State and City1,...,City5
c1 = largest_cities[['State', 'City1']]
c2 = largest_cities[['State', 'City2']]
c3 = largest_cities[['State', 'City3']]
c4 = largest_cities[['State', 'City4']]
c5 = largest_cities[['State', 'City5']]
```

```python
# change City1,...,City5 to City
c1.columns = ['State', 'City']
c2.columns = ['State', 'City']
c3.columns = ['State', 'City']
c4.columns = ['State', 'City']
c5.columns = ['State', 'City']
```

```python
c1.head()
```

```python
df = pd.concat([c1, c2, c3, c4, c5], axis=0)
df
```

```python
# remove rows with NaN values in City column
df.dropna(subset=['City'], inplace=True)
```

```python
df.shape
```

```python
# remove digit at the end of City values, if any
# eg. Honolulu1 to Honolulu
df['City'] = df['City'].str.replace(r'\d$', '', regex=True)
```

```python
df['State'] = df['State'].str.replace(r'\d$', '', regex=True)
```
