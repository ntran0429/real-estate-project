```python
import os
from dotenv import load_dotenv
import requests
import pandas as pd
import matplotlib as plt
import json
from fredapi import Fred
```

```python
load_dotenv()
fred_api_key = os.getenv("FRED_API_KEY")
fred = Fred(api_key=fred_api_key)
```

```python
df_rpp = fred.search('Regional Price Parities: All Items')
# df_inc = df_inc.query('units == ""')

df_rpp = df_rpp.loc[df_rpp['title'].str.contains('Regional Price Parities: All Items for')]
```

```python
df_rpp.tail()
```

```python
rpp_result = df_rpp[~df_rpp['title'].str.endswith('(MSA)')]
```

```python
rpp_result.shape
```

```python
all_states = [
    'AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA',
    'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK',
    'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY'
]

# Convert your series index to a list
existing_states = list(rpp_result.index.str[:2])

# Find the missing states
missing_states = set(all_states) - set(existing_states)

# Display the missing states
print("Missing States:", missing_states)
```

```python
rpp_result['title']
```

```python
all_result =[]

for id in rpp_result.index:
    result = fred.get_series(id)
    result = result.to_frame(name = id)
    all_result.append(result)
```

```python
all_result[1]
```

```python
rpp_state = pd.concat(all_result, axis=1)
```

```python
rpp_state.tail()
```

```python
id_to_state = df_rpp['title'].str.replace('Regional Price Parities: All Items for ','').to_dict()
```

```python
rpp_state.columns = [id_to_state[d] for d in rpp_state.columns]
```
