```python
import os
from dotenv import load_dotenv
import requests
import pandas as pd
import matplotlib as plt
import json
from fredapi import Fred
```

```python
load_dotenv()
fred_api_key = os.getenv("FRED_API_KEY")
fred = Fred(api_key=fred_api_key)
```

```python
df_pop = fred.search('Resident Population')
# df_inc = df_inc.query('units == "Dollars"')

df_pop = df_pop.loc[df_pop['title'].str.contains('Resident Population')]
```

```python
df_pop.shape
```

```python
df_pop.head()
```

```python
pop_result = df_pop[df_pop['id'].str.endswith('POP')]
```

```python
pop_result.id.unique()
```

```python
pop_result = pop_result[pop_result['id'].str.len() == 5]
```

```python
pop_result.shape
```

```python
pop_result.id.unique()
```

```python
pop_result = pop_result[~pop_result['id'].str.contains('D5POP')]
```

```python
all_result =[]

for id in pop_result.index:
    result = fred.get_series(id)
    result = result.to_frame(name = id)
    all_result.append(result)
```

```python
all_result[1]
# Thousands of Persons
```

```python
pop_state = pd.concat(all_result, axis=1)
```

```python
pop_state.tail()
```

```python
# create a mapping dict to change column names using list comprehension
id_to_state = df_pop['title'].str.replace('Resident Population in ','').to_dict()
```

```python
pop_state.columns
```

```python
pop_state.columns = [id_to_state[d] for d in pop_state.columns]
```
