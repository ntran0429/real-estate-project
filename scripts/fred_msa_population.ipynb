```python
import os
from dotenv import load_dotenv
import requests
import pandas as pd
import matplotlib as plt
import json
from fredapi import Fred
```

```python
load_dotenv()
fred_api_key = os.getenv("FRED_API_KEY")
fred = Fred(api_key=fred_api_key)
```

```python
df_msa_pop = fred.search('Resident Population')
# df_inc = df_inc.query('units == ""')

df_msa_pop = df_msa_pop.loc[df_msa_pop['title'].str.contains('Resident Population')]
```

```python
df_msa_pop['seasonal_adjustment'].value_counts()
```

```python
df_msa_pop
```

```python
msa_pop_result = df_msa_pop[df_msa_pop['title'].str.contains('(MSA)')]
```

```python
msa_pop_result.shape
```

```python
msa_pop_result
```

```python
all_result =[]

for id in msa_pop_result.index:
    result = fred.get_series(id)
    result = result.to_frame(name = id)
    all_result.append(result)
```

```python
all_result[1]
# Thousands of Persons
```

```python
pop_msa = pd.concat(all_result, axis=1)
```

```python
pop_msa.head()
```

```python
# keep only the first 23 rows
pop_msa = pop_msa.iloc[:23,]
```

```python
msa_pop_result.title
```

```python
id_to_state = msa_pop_result['title'].str.replace('Resident Population in ','').to_dict()
```

```python
pop_msa.columns = [id_to_state[d] for d in pop_msa.columns]
```
